```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from matplotlib import colors
#read the dataset
dataset =pd.read_csv("C:\\Users\\Noel\\Downloads\\iris.csv")
#display the columns in dataset
dataset.columns
#calculate the mean of each columns

dataset.mean()
```

```python
#using histograph
plt.hist(dataset['SL'])

plt.xlabel("SL")
plt.ylabel("freequency")
plt.title('histograph')
plt.show()
```
