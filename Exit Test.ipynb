### Preprocessing

#### Label Encoding

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
for i in df.columns:
    if df[i].dtype == 'object':
        df[i] = le.fit_transform(df[i])
```

```python
df.columns
```

```python
df.shape
```

```python
x=df.drop(['selection'],axis=1)
```

```python
y=df['selection']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,random_state =101,test_size=0.33)
```

## Modelling

### 1)Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression()
log_model = log_model.fit(x_train,y_train)
y_pred = log_model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,recall_score,f1_score,precision_score
print('accuracy is=:',accuracy_score(y_test,y_pred))
```

### 2)KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
metric_k = []
neighbors = np.arange(3,15)
```

```python
for k in neighbors:
    classifier = KNeighborsClassifier(n_neighbors = k,metric = 'euclidean')
    classifier.fit(x_train,y_train)
y_prediction = classifier.predict(x_test)
acc = accuracy_score(y_test,y_prediction)
metric_k.append(acc)
metric_k
```

```python
classifier = KNeighborsClassifier(n_neighbors = 5,metric = 'euclidean')
classifier.fit(x_train,y_train)
y_prediction = classifier.predict(x_test)
print('accuracy is=:',accuracy_score(y_test,y_prediction))
```

```python
confusion_matrix(y_test,y_prediction)
```

### 3)Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
```

```python
y_pred2 = rf.predict(x_test)
cm = confusion_matrix(y_test, y_pred2)
print(cm)
print('Accuracy is',accuracy_score(y_test, y_pred2))
print('Recall is',recall_score(y_test, y_pred2,average='micro'))
print('Precision is',precision_score(y_test, y_pred2,average='micro'))
```

### 4)Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

```python
y_pred3 = dt.predict(x_test)
cm = confusion_matrix(y_test, y_pred3)
print(cm)
print('Accuracy is',accuracy_score(y_test, y_pred3))
print('Recall is',recall_score(y_test, y_pred3,average='micro'))
print('Precision is',precision_score(y_test, y_pred3,average='micro'))
```

### 5)Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingClassifier
gb=GradientBoostingClassifier()
gb.fit(x_train,y_train)
```

```python
y_pred4 = gb.predict(x_test)
cm = confusion_matrix(y_test, y_pred4)
print(cm)
print('Accuracy is',accuracy_score(y_test, y_pred4))
print('Recall is',recall_score(y_test, y_pred4,average='micro'))
print('Precision is',precision_score(y_test, y_pred4,average='micro'))
```
